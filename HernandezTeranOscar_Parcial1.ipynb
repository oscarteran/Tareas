```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f1 = lambda x: ((np.pi*(np.exp(1)-1))/(2))*(np.cos((np.pi*x)/(2)))-np.exp(1)**x

xmin = -2
xmax = 2
x = np.linspace(xmin, xmax, 100)

plt.plot(x,f1(x))
plt.plot([xmin, xmax], [0,0], 'r-', lw=1)
plt.grid()
```

```python
f2 = lambda x: np.tan(x)+(1/np.cos(x))**2*(x-1)+np.sin(np.pi*x)+np.pi*x*np.cos(np.pi*x)

xmin = -1
xmax = 1
x = np.linspace(xmin, xmax, 100)

plt.plot(x,f2(x))
plt.plot([xmin, xmax], [0,0], 'r-', lw=1)
plt.grid()
```

```python
f3 = lambda x: np.sin(np.pi*x)+np.pi*x*np.cos(np.pi*x)-((x-2)/x)-np.log(x)

xmin = 0.1
xmax = 2
x = np.linspace(xmin, xmax, 100)

plt.plot(x,f3(x))
plt.plot([xmin, xmax], [0,0], 'r-', lw=1)
plt.grid()
```

```python
f4 = lambda x: np.cos(np.pi*x)*np.log(x+2)*(x-2)+(((x+2)/(x-2))+np.log(x+2))*np.sin(x)

xmin = -1
xmax = 1
x = np.linspace(xmin, xmax, 100)

plt.plot(x,f4(x))
plt.plot([xmin, xmax], [0,0], 'r-', lw=1)
plt.grid()
```

```python
# Metodo a
n = 21
suma = np.zeros(n)
for i in range(n):
    suma[i] = ((-1)**i*5**i)/np.math.factorial(i)
print("Valor a estimar: ", 0.00673)
print(np.sum(suma))

# Metodo b
n2 = 14
suma2 = np.zeros(n2)
for j in range(n2):
    suma2[j] = (5**j)/(np.math.factorial(j))
res = 1/(np.sum(suma2))
print(res)
```
